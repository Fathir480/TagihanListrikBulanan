```python
def input_kwh(nama):
    while True:
        kwh = float(input(f'Masukkan pemakaian kWh untuk {nama}: '))
        if not isinstance(kwh, float):
            print('Input tidak valid, masukkan angka')
            continue
        if kwh < 0:
            print('kWh tidak boleh negatif')
            continue
        return kwh

def tentukan_golongan(kwh):
    if kwh <= 100:
        return "R-1"
    elif kwh <= 300:
        return "R-2"
    else:
        return "R-3"

def hitung_tagihan(kwh, golongan):
    if golongan == "R-1":
        tarif = 1352
        abodemen = 15000
    elif golongan == "R-2":
        tarif = 1699
        abodemen = 25000
    else:
        tarif = 2082
        abodemen = 50000

    total = (kwh*tarif) + abodemen
    return total

def cek_status(tagihan, golongan):
    if golongan == 'R-1':
        batas_rendah = 100_000
        batas_normal = 150_000
    elif golongan == 'R-2':
        batas_rendah = 200_000
        batas_normal = 400_000
    else:
        batas_rendah = 500_000
        batas_normal = 700_000

    if tagihan < batas_rendah:
        return "Rendah"
    elif tagihan <= batas_normal:
        return "Normal"
    else:
        return "Tinggi"

def cetak_struk(nama, kwh, golongan, tagihan, status):
    print('----------------------')
    print(f'Nama      : {nama}')
    print(f'Pemakaian : {kwh} kWh')
    print(f'Golongan  : {golongan}')
    print(f'Tagihan   : Rp {tagihan:,.0f}')
    print(f'Status    : {status}')
    print('----------------------')

def cetak_rekap(data_warga):
    total = 0
    for nama, tagihan in data_warga:
        total += tagihan

    rata_rata = total / len(data_warga)

    print('--- REKAP BULANAN ---')
    print(f'Jumlah warga  : {len(data_warga)}')
    print(f'Total tagihan : Rp {total:,.0f}')
    print(f'Rata-rata     : Rp {rata_rata:,.0f}')

def judul():
    print('=======================================')
    print('= KALKULATOR TAGIHAN LISTRIK KOPERASI =')
    print('=======================================')

def main():
    print(judul())
    data_warga = []

    while True:
        print('\n=======================================')
        nama = input("Nama warga (atau 'selesai'): ")

        if nama.lower() == 'selesai':
            break

        kwh = input_kwh(nama)
        golongan = tentukan_golongan(kwh)
        tagihan = hitung_tagihan(kwh, golongan)
        status = cek_status(tagihan, golongan)

        cetak_struk(nama, kwh, golongan, tagihan, status)

        data_warga.append((nama,tagihan))
        
        if data_warga:
            cetak_rekap(data_warga)
        else:
            print('Tidak ada data yang dimasukkan')
main()
```
